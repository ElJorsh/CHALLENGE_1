### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
df = pd.concat([tienda, tienda2, tienda3, tienda4])
df.head()
```

```python

fact_tienda1 = (tienda["Precio"] + tienda["Costo de envío"]).sum()
print(f" Facturación Tienda 1: ${fact_tienda1:,.0f}")


fact_tienda2 = (tienda2["Precio"] + tienda2["Costo de envío"]).sum()
print(f" Facturación Tienda 2: ${fact_tienda2:,.0f}")


fact_tienda3 = (tienda3["Precio"] + tienda3["Costo de envío"]).sum()
print(f" Facturación Tienda 3: ${fact_tienda3:,.0f}")


fact_tienda4 = (tienda4["Precio"] + tienda4["Costo de envío"]).sum()
print(f" Facturación Tienda 4: ${fact_tienda4:,.0f}")
```

```python
df = pd.concat([tienda, tienda2, tienda3, tienda4])
```

```python
import matplotlib.pyplot as plt

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
facturacion = [fact_tienda1, fact_tienda2, fact_tienda3, fact_tienda4]

plt.figure()
plt.bar(tiendas, facturacion)
plt.xlabel("Tiendas")
plt.ylabel("Facturación")
plt.title("Facturación por Tienda")

for i, valor in enumerate(facturacion):
    plt.text(i, valor, f"{valor:,.0f}", ha="center", va="bottom")

plt.show()
```

# 2. Ventas por categoría

```python
Categorias_popularest1 = tienda["Categoría del Producto"].value_counts()
print("\n Categorias_populares Tienda 1")
print(Categorias_popularest1)

Categorias_popularest2 = tienda2["Categoría del Producto"].value_counts()
print("\n Categorias_populares Tienda 2")
print(Categorias_popularest2)

Categorias_popularest3 = tienda3["Categoría del Producto"].value_counts()
print("\n Categorias_populares Tienda 3")
print(Categorias_popularest3)
Categorias_popularest4 = tienda4["Categoría del Producto"].value_counts()
print("\n Categorias_populares Tienda 4")
print(Categorias_popularest4)
```

# 3. Calificación promedio de la tienda


```python
Calificacion_promedio1 = tienda["Calificación"].mean()
print(f" Calificación promedio Tienda 1: {Calificacion_promedio1:,.2f}")

Calificacion_promedio2 = tienda2["Calificación"].mean()
print(f" Calificación promedio Tienda 2: {Calificacion_promedio2:,.2f}")

Calificacion_promedio3 = tienda3["Calificación"].mean()
print(f" Calificación promedio Tienda 3: {Calificacion_promedio3:,.2f}")

Calificacion_promedio4 = tienda4["Calificación"].mean()
print(f" Calificación promedio Tienda 4: {Calificacion_promedio4:,.2f}")
```

```python

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones = [
    Calificacion_promedio1,
    Calificacion_promedio2,
    Calificacion_promedio3,
    Calificacion_promedio4]

plt.figure()
plt.plot(tiendas, calificaciones, marker='o')
plt.title("Comparación de Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)

# Etiquetas de puntos
for i, valor in enumerate(calificaciones):
    plt.text(i, valor, f"{valor:.2f}", ha="center", va="bottom")

plt.show()
```

# 4. Productos más y menos vendidos

```python
mas_vendidost1 = tienda["Producto"].value_counts()
print("\n Productos mas vendidos Tienda 1")
print(mas_vendidost1.head(5))

menos_vendidost1 = tienda["Producto"].value_counts()
print("\n Productos menos vendidos Tienda 1")
print(menos_vendidost1.tail(5))

mas_vendidost2 = tienda2["Producto"].value_counts()
print("\n Productos mas vendidos Tienda 2")
print(mas_vendidost2.head(5))

menos_vendidost2 = tienda2["Producto"].value_counts()
print("\n Productos menos vendidos Tienda 2")
print(menos_vendidost2.tail(5))

mas_vendidost3 = tienda3["Producto"].value_counts()
print("\n Productos mas vendidos Tienda 3")
print(mas_vendidost3.head(5))

menos_vendidost3 = tienda3["Producto"].value_counts()
print("\n Productos menos vendidos Tienda 3")
print(menos_vendidost3.tail(5))

mas_vendidost4 = tienda4["Producto"].value_counts()
print("\n Productos mas vendidos Tienda 4")
print(mas_vendidost4.head(5))

menos_vendidost4 = tienda4["Producto"].value_counts()
print("\n Productos menos vendidos Tienda 4")
print(menos_vendidost4.tail(5))
```

# 5. Envío promedio por tienda

```python
envio_promediot1 = tienda["Costo de envío"].mean()
print(f" Envío promedio Tienda 1: ${envio_promediot1:,.2f}")

envio_promediot2 = tienda2["Costo de envío"].mean()
print(f" Envío promedio Tienda 2: ${envio_promediot2:,.2f}")

envio_promediot3 = tienda3["Costo de envío"].mean()
print(f" Envío promedio Tienda 3: ${envio_promediot3:,.2f}")

envio_promediot4 = tienda4["Costo de envío"].mean()
print(f" Envío promedio Tienda 4: ${envio_promediot4:,.2f}")
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
envio_promedio = [envio_promediot1,
    envio_promediot2,
    envio_promediot3,
    envio_promediot4]

# Crear gráfico de torta
plt.figure()
plt.pie(
    envio_promedio,
    labels=tiendas,
    autopct="%.1f%%",
    startangle=90)
plt.title("Distribución del Envío Promedio por Tienda")

plt.show()
```
